```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
df.describe()
```

```python
df.rename(columns={"Life expectancy at birth (years)": "Life Expectancy"}, inplace=True)
```

```python
us_df = df[df['Country'] == "United States of America"]
german_df = df[df['Country'] == 'Germany']
mexico_df = df[df['Country'] == 'Mexico']
chile_df = df[df['Country'] == 'Chile']
china_df = df[df['Country'] == 'China']
zimb_df = df[df['Country'] == 'Zimbabwe']
```

```python
countries = df['Country'].unique()
print(countries)
```

```python
for country in countries:
  plt.title(f"GDP(Trillions of USD) of {country} over Time")
  sns.lineplot(data=df[df['Country'] == country], x='Year', y='GDP')
  plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country', alpha=0.6)
plt.legend(loc='upper left', fontsize=7.5)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(data=df, x='Year', y='Life Expectancy', hue='Country', alpha=0.6)
plt.legend(loc='lower right', fontsize=7.5)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=df, x='Life Expectancy', y='GDP', hue='Country')
plt.show()
```

```python
for country in countries:
  plt.figure(figsize=(4, 3))
  plt.title(f"Life Expectancy of {country} compared to GDP(Trillions of USD")
  sns.scatterplot(data=df[df['Country'] == country], x='Life Expectancy', y='GDP')
  plt.show()
```
